## Compare image statistics (galaxy image v.s. celebra)

```python
def display_img_stats(image):
    img_npy = np.array(image)
    print(f'(min, max) = ({img_npy.min()}, {img_npy.max()})')
    print(f'(mean, std) = ({img_npy.mean():.3f}, {img_npy.std():.3f})')
    
def display_rgb_stats(image):
    r, g, b = image.split()
    print('R')
    display_img_stats(r)
    print('G')
    display_img_stats(g)
    print('B')
    display_img_stats(b)
```

```python
img_celebra = Image.open(os.path.join(dir_image, 'celeba_1.jpg'))
img_galaxy  = Image.open(os.path.join(dir_image, 'gz2_16.jpg'))

fig, ax = plt.subplots(1, 2)
ax[0].imshow(img_celebra)
ax[1].imshow(img_galaxy)

for i in range(2):
    ax[i].axis('off')
```

```python
display_rgb_stats(img_celebra)
```

```python
display_rgb_stats(img_galaxy)
```
